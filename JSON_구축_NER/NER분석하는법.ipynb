```python
#에트리 개체명 인식 분석하자
#-*- coding:utf-8 -*-
import urllib3 #기능 1은 인터넷 접속을 해서 http 특정 주소를 넘겨주고 접속ㅇ함.
import json
```

```python
openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU_spoken"
```

```python
accessKey = "204a0e5d-baaa-4144-9c18-d4b44f69e14b" #204a0e5d-baaa-4144-9c18-d4b44f69e14b내가발급받은키임
analysisCode = "ner" #어떤 분석을 할 것인지? 이건 
text = ""
```

```python
text = "근데 형 이 데이터 라벨러라는 게 아직까지는 이렇게 정식계약도 근로계약도 아니고 용역계약을 하기 때문에 짬짬이 하는 아르바이트 개념으로 보는 게 더 편할 것 같어 이쪽 시장이 커지면 분명 더 좋은 메리트이겠지만 아직까지는 그래 보여"
```

```python
requestJson = {
    "access_key": accessKey,#내 변수
    "argument": {
        "text": text,
        "analysis_code": analysisCode
    }
}
```

```python
http = urllib3.PoolManager()
response = http.request(
    "POST",
    openApiURL,
    headers={"Content-Type": "application/json; charset=UTF-8"},
    body=json.dumps(requestJson)
)
print(str(response.data,"utf-8"))
```

```python
result = json.loads(response.data) #json을 담음
print(result)
```

```python
NERS = [] #리스트 생성
#텍스트와 타입을 쌍으로 튜플로 만들어서 넣어줄게 
for data in sents:
    for ne in data['NE']:
        NERS.append( (ne['type'],ne['text']) )
```

```python
#각각의 태그에서 몇 번의 빈도가 나오는지 디셔너리를 활용해서 알아보자
```

```python
tag_dic = {}
for tag, word in NERS:
    tag_dic[tag] = tag_dic.get(tag, 0)+1 #tag_dic딕셔너리에 [tag]이름을 가지는 것을 가지고와라 아니면 0을 넣어라 그리고 포문을 돌며 +1씩 쌓는다
```

```python
tag_dic
```

```python
tag_list = []
for tag, freq in tag_dic.items():
    tag_list.append((freq,tag))
tag_list.sort(reverse=True)
```
