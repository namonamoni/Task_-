```python
#에트리 개체명 인식 분석하자
#-*- coding:utf-8 -*-
import urllib3
import json
```

```python
openApiURL = "http://aiopen.etri.re.kr:8000/WiseNLU_spoken" 
```

```python
accessKey = "204a0e5d-baaa-4144-9c18-d4b44f69e14b" #204a0e5d-baaa-4144-9c18-d4b44f69e14b
analysisCode = "ner" #어떤 분석을 할 것인지? 이건 
text = ""
```

```python
import pandas as pd
```

```python
df = pd.read_excel('1994남10대.xlsx') #10,000글자씩 자른 10덩이를 Etri에 업로드
```

```python
from tqdm.notebook import tqdm
```

```python
etri_data = []
for data in tqdm(df['문장']):
    text = data

    requestJson = {
        "access_key": accessKey,#내 변수
        "argument": {
            "text": text,
            "analysis_code": analysisCode
        }
    }
    
    http = urllib3.PoolManager()
    response = http.request(
        "POST",
        openApiURL,
        headers={"Content-Type": "application/json; charset=UTF-8"},
        body=json.dumps(requestJson)
    )
    result = json.loads(response.data)
    sents = result['return_object']['sentence']
    
    for item in sents:
        etri_data.append(item)
        
```

```python
NERS = [] #리스트 생성
#텍스트와 타입을 쌍으로 튜플로 만들어서 넣어줄게 
for data in sents:
    for ne in data['NE']:
        NERS.append( (ne['type'],ne['text']) )
```

```python
#각 태그에서 몇 번의 빈도가 나오는지 디셔너리를 활용해서 알아보자
```

```python
tag_dic = {}
for tag, word in NERS:
    tag_dic[tag] = tag_dic.get(tag, 0)+1
```
